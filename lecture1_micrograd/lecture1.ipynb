```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits import mplot3d 
```

```python
def f(x):
    return 3*x**2 - 4*x + 5
```

```python
f(3.0)
```

```python
xs=np.arange(-2,5,0.25)
print(len(xs))
ys=f(xs)
plt.plot(xs,ys)
```

```python
x=2/3
h=0.001

dx=(f(x+h)-f(x))/h
dx
```

```python
#partial derivative

a=2
b=3
c=10
f=a*b**2-c

d1=a*b**2-c
c+=h
d2=a*b**2-c
print("d1",d1)
print("d2",d2)
print("slope",(d2-d1)/h)
```

```python
m=np.arange(2,6)
n=np.arange(3,7)

m, n=np.meshgrid(m,n)
o=m*n**2
print(o)
fig=plt.figure(figsize=(20,40))

ax = plt.axes(projection='3d')

# Choose a plot type: Surface, Wireframe, or Contour
surf = ax.plot_surface(m, n, o, cmap='viridis', edgecolor='none')

# Add customization
ax.set_title('3D Surface Plot of FXN')
ax.set_xlabel('m axis')
ax.set_ylabel('n axis')
ax.set_zlabel('0 axis')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5) # Add a color bar

plt.show()
```
